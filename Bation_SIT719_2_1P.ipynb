Step 1. Import pandas and numpy libraries

```python
import pandas as pd
import numpy as np
```

Step 2. Import the popular 'iris' data sets'

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url)
iris.head()
```

Step 3. Add column headers


1.   sepal_length
2.   sepal_width
3.   petal_length
4.   petal_width
5.   class



```python
iris.columns = ['sepal_length','sepal_width','petal_length','petal_width','class']
iris.head()
```

Step 4. Check if there are missing values in dataframe

```python
iris.isnull().sum()
```

Step 5. Write a code to set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
iris.loc[10:29, 'petal_length'] = np.nan
print(iris.loc[10:29])
```

Step 6. Check if there are missing values in dataframe again

```python
iris.isnull().sum()
```

Step 7. Substitute NaN values to 10.0

```python
iris = iris.fillna(10.0)
print(iris.loc[10:29])
```

Python provides a wide range of libraries that can be used in data analysis, manipulation and visualization. Below is a list of libraries commonly used:

*   Pandas library is useful for data cleaning and analysis
*   Numpy library is useful for performing computations and math operations
*   Matplotlib library is useful for data visualization such as plots & graphs

Two of the most common functionalities used with regards to data manipulation are (1) checking of null values in the dataFrame, and (2) filtering rows based on some condition.

Below is an example of how these 2 functionalities are used and also how matplotlib is used to visualize data.



```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Student': ['April', 'Danuth', 'Jom', 'David', 'Ice'],
    'Age': [25, np.nan, 30, 22, 28],
    'Grade': ['P', 'HD', np.nan, 'HD', 'C']
}
df = pd.DataFrame(data)

# Example 1: Checking for Missing Values
print("Missing values in each column:")
print(df.isnull().sum())

# Example 2: Filtering Employees with Grade == 'HD'
hd_students = df[df['Grade'] == 'HD']
print("\nStudents with an HD grade")
print(hd_students)

# Using matplotlib to visualize data
grade_counts = df['Grade'].value_counts(dropna=True)
plt.figure(figsize=(3, 3))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%')
plt.title("\nDistribution of Grades")
plt.show()
```
